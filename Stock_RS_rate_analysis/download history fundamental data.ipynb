### 2. 定義分數標準

```python
# 碩分數
"""
稅後淨利率yoy :縮寫 IAT_R_YoY
稅後淨利率yoy兩季平均 : 縮寫 IAT_YoY_AVG_2Q
稅後淨利率yoy兩季平均的增減 : 縮寫 IAT_YoY_AVG_2Q_Diff
毛利率yoy : 縮寫 GP_R_YoY
毛利率yoy兩季平均 : 縮寫 GP_R_YoY_AVG_2Q
毛利率yoy兩季平均的增減 : 縮寫 GP_R_YoY_AVG_2Q_Diff
營業利益率yoy : 縮寫 OI_R_YoY
營業利益率yoy兩季平均 : 縮寫 OI_YoY_AVG_2Q
營業利益率yoy兩季平均的增減 : 縮寫 OI_YoY_AVG_2Q_Diff
EPSyoy : 縮寫 EPS_YoY
EPSyoy兩季平均 : 縮寫 EPS_YoY_AVG_2Q
EPSyoy兩季平均的增減 : 縮寫 EPS_YoY_AVG_2Q_Diff
月營收yoy : 縮寫 M_Rev_YoY
月營收yoy兩月平均 : 縮寫 M_Rev_YoY_AVG_2M
月營收yoy兩月平均的增減 : 縮寫 M_Rev_YoY_AVG_2M_Diff
"""

# 敏分數
"""
    稅後淨利率年增率 IAT_R_YoY 評分標準：
    AA: IAT_R_YoY 最近三季維持正數 且 IAT_R_YoY 最近一季增長
    A: IAT_R_YoY 最近兩季正數 且 最近兩季 IAT_R_YoY > -50%
    BB: IAT_R_YoY最近兩季正數 且 最近一季 IAT_R_YoY  < -50%
    B: IAT_R_YoY 上一季出現負數
    C: IAT_R_YoY 最近一季出現負數
    備註: 若同時符合兩個條件，取低分為準。
"""
"""
    EPS 評分標準：
    AA: EPS 最近四季累積超過 5 元
    A: EPS 最近四季累積 3-5 元
    BB: EPS 最近四季累積 1-3 元
    B: EPS 最近四季累積超過 0 元 or 最近一季出現虧損
    C: EPS 最近四季累積虧損
    備註: 若同時符合兩個條件，取低分為準。
"""
"""
    月營收年增率 M_Rev_YoY 評分標準：
    AA: M_Rev_YoY 最近六個月皆為正數 且  M_Rev_YoY 最近六個月平均超過25% 且 最近一個月 M_Rev_YoY >= 上一個月 M_Rev_YoY
    A: M_Rev_YoY 最近六個月皆為正數 且 M_Rev_YoY 最近六個月平均在 10-25% 且 最近一個月 M_Rev_YoY >= 上一個月 M_Rev_YoY
       or M_Rev_YoY 最近六個月皆為正數 且  M_Rev_YoY 最近六個月平均超過25% 且 (最近一個月 M_Rev_YoY - 上一個月 M_Rev_YoY)/abs(上一個月 M_Rev_YoY) > 0.5 
    BB: M_Rev_YoY 最近六個月內曾經出現單月負成長 or 無法被評為其他等級
    B: M_Rev_YoY 最近六個月平均 > 0 且 這個月 M_Rev_YoY < 上個月 M_Rev_YoY < 上上個月 M_Rev_YoY 
       or M_Rev_YoY的MoM 最近四個月 出現三次負數 
    C: 最近一個月為負數 or 最近六個月平均為負數
    備註: 若同時符合兩個條件，取低分為準。
"""
"""
    存貨周轉率 IT 評分標準：
    AA: IT_QoQ 最近四季皆 > -20%，IT 最近四季平均在 1.5 次以上
    A: IT_QoQ 最近四季皆 > -20%，平均在 1.5 次以下
    BB: IT_QoQ 最近四季出現連續兩季負值，累積跌幅 20% 以上
    B: IT_QoQ 最近四季出現單季 < -20%
    C: IT_QoQ 最近一季 < -20% 
    備註: 若同時符合兩個條件，取低分為準。
"""
"""
    自由現金流 FCF 評分標準：
    AA: FCF 連續六季出現正數
    A: FCF 最近六季累積正數 且 FCF 最近四季累積正數
    BB: FCF 最近六季累積負數 且 FCF 最近四季累積正數
    B: FCF 最近六季累積正數 且 FCF 最近四季累積負數
    C: FCF 最近六季累積負數 且 FCF 最近四季累積負數
    備註: 若同時符合兩個條件，取低分為準。
"""
```

### 3. 計算月營收分數資訊(年增率...)

```python
import pandas as pd
import os

# 讀取來源檔案與範本檔案
file_path = r'C:\Users\User\Desktop\stockchoose\history_month_revenue_data'
file_list = os.listdir(file_path)
source_file = [file for file in file_list if file.endswith('.xlsx')]
template_file = r'C:\Users\User\Desktop\stockchoose\history_month_revenue_data\範本.xlsx'
output_file = r'C:\Users\User\Desktop\stockchoose\history_month_revenue_data\monthly_revenue_data.xlsx'

# 讀取範本檔案
template_df = pd.read_excel(template_file, header=0)

# 初始化結果 DataFrame
converted_df = pd.DataFrame()

# 依據每個日期進行計算
for file in source_file[:]:
    source_df = pd.read_excel(fr'{file_path}\{file}', index_col=0, header=0)
    converted_df = pd.DataFrame()
    try:
        for date in source_df['update_date'].unique():
            date_df = source_df[source_df['update_date'] == date]
            
            # 計算財報資訊
            M_Rev = date_df.loc[:, 'revenue'].sum()
            
            # 取得時間資訊
            stock_id = date_df['stock_id'].iloc[0]
            year = date_df['year'].iloc[0]
            month = date_df['month'].iloc[0]
            day = date_df['day'].iloc[0]
            update_date = date_df['update_date'].iloc[0]
            Rev_month = date_df['revenue_month'].iloc[0]
            Rev_quarter = date_df['revenue_quarter'].iloc[0]
            Rev_year = date_df['revenue_year'].iloc[0]
            
            
            
            # 將計算結果添加到結果 DataFrame
            converted_df = pd.concat([converted_df, pd.DataFrame({
                'ID': stock_id,
                'Y': year,
                'Rev_M': Rev_month,
                'Rev_Q': Rev_quarter,
                'Rev_Y': Rev_year,
                'M': month,
                'D': day,
                'M_Rev': M_Rev,
                'update_date': update_date
            }, index=[0])], ignore_index=True)
            # 計算年增率
            converted_df['M_Rev_YoY'] = (converted_df['M_Rev'] - converted_df['M_Rev'].shift(12))/abs(converted_df['M_Rev'].shift(12))
            converted_df['M_Rev_MoM'] = (converted_df['M_Rev'] - converted_df['M_Rev'].shift(1))/abs(converted_df['M_Rev'].shift(1))
            converted_df['M_Rev_YoY_MIN_6M'] = converted_df['M_Rev_YoY'].rolling(6).min()
            converted_df['M_Rev_YoY_AVG_6M'] = converted_df['M_Rev_YoY'].rolling(6).mean()
            converted_df['M_Rev_YoY_MoM'] = (converted_df['M_Rev_YoY'] - converted_df['M_Rev_YoY'].shift(1))/abs(converted_df['M_Rev_YoY'].shift(1))
            converted_df['M_Rev_YoY_AVG_2M'] = converted_df['M_Rev_YoY'].rolling(2).mean()
            converted_df['M_Rev_YoY_AVG_2M_Diff'] = converted_df['M_Rev_YoY_AVG_2M'] - converted_df['M_Rev_YoY_AVG_2M'].shift(1)

    except Exception as e:
        print(file, e)
        continue
    
    template_df = pd.concat([template_df, converted_df], ignore_index=True, join='outer')
    print(file, "success")

# 儲存結果
template_df.to_excel(output_file, index=False)
```

```python
import pandas as pd
import os

# 讀取來源檔案與範本檔案
file_path = r'C:\Users\User\Desktop\stockchoose\history_financial_statement_data'
file_list = os.listdir(file_path)
source_file = [file for file in file_list if file.endswith('.xlsx')]
template_file = r'C:\Users\User\Desktop\stockchoose\history_financial_statement_data\範本.xlsx'
output_file = r'C:\Users\User\Desktop\stockchoose\history_financial_statement_data\financial_statement_data.xlsx'

# 讀取範本檔案
template_df = pd.read_excel(template_file, header=0)

# 初始化結果 DataFrame
converted_df = pd.DataFrame()

# 依據每個日期進行計算
for file in source_file[:]:
    source_df = pd.read_excel(fr'{file_path}\{file}', index_col=0, header=0)
    converted_df = pd.DataFrame()
    try:
        for date in source_df['update_date'].unique():
            date_df = source_df[source_df['update_date'] == date]
            
            # 計算財報資訊
            IAT = date_df.loc[date_df['type'] == 'IncomeAfterTaxes', 'value'].sum()
            Q_Rev = date_df.loc[date_df['type'] == 'Revenue', 'value'].sum()
            IAT_R = (IAT / Q_Rev) * 100 if Q_Rev != 0 else 0
            OI = date_df.loc[date_df['type'] == 'OperatingIncome', 'value'].sum()
            OI_R = (OI / Q_Rev) * 100 if Q_Rev != 0 else 0
            GP = date_df.loc[date_df['type'] == 'GrossProfit', 'value'].sum()
            GP_R = (GP / Q_Rev) * 100 if Q_Rev != 0 else 0
            EPS = date_df.loc[date_df['type'] == 'EPS', 'value'].mean()
            

            # 取得時間資訊
            stock_id = date_df['stock_id'].iloc[0]
            year = date_df['year'].iloc[0]
            month = date_df['month'].iloc[0]
            day = date_df['day'].iloc[0]
            update_date = date_df['update_date'].iloc[0]
            Rev_quarter = date_df['revenue_quarter'].iloc[0]
            Rev_year = date_df['revenue_year'].iloc[0]
            
            
            
            # 將計算結果添加到結果 DataFrame
            converted_df = pd.concat([converted_df, pd.DataFrame({
                'ID': stock_id,
                'Y': year,
                'Rev_Q': Rev_quarter,
                'Rev_Y': Rev_year,
                'M': month,
                'D': day,
                'IAT': IAT,
                'IAT_R': IAT_R,
                'Q_Rev': Q_Rev,
                'EPS': EPS,
                'OI': OI,
                'OI_R': OI_R,
                'GP': GP,
                'GP_R': GP_R,
                'update_date': update_date
            }, index=[0])], ignore_index=True)
            # 計算年增率
            converted_df['Q_Rev_YoY'] = (converted_df['Q_Rev'] - converted_df['Q_Rev'].shift(4)) / abs(converted_df['Q_Rev'].shift(4)) * 100 if len(converted_df) > 3 else 0
            converted_df['IAT_R_YoY'] = (converted_df['IAT_R'] - converted_df['IAT_R'].shift(4)) / abs(converted_df['IAT_R'].shift(4)) * 100 if len(converted_df) > 3 else 0
            converted_df['OI_R_YoY'] = (converted_df['OI_R'] - converted_df['OI_R'].shift(4)) / abs(converted_df['OI_R'].shift(4)) * 100 if len(converted_df) > 3 else 0
            converted_df['EPS_YoY'] = (converted_df['EPS'] - converted_df['EPS'].shift(4)) / abs(converted_df['EPS'].shift(4)) * 100 if len(converted_df) > 3 else 0
            converted_df['GP_R_YoY'] = (converted_df['GP_R'] - converted_df['GP_R'].shift(4)) / abs(converted_df['GP_R'].shift(4)) * 100 if len(converted_df) > 3 else 0

            # 計算季增率
            converted_df['Q_Rev_QoQ'] = (converted_df['Q_Rev'] - converted_df['Q_Rev'].shift(1)) / abs(converted_df['Q_Rev'].shift(1)) * 100 if len(converted_df) > 0 else 0
            converted_df['IAT_R_QoQ'] = (converted_df['IAT_R'] - converted_df['IAT_R'].shift(1)) / abs(converted_df['IAT_R'].shift(1)) * 100 if len(converted_df) > 0 else 0
            converted_df['OI_R_QoQ'] = (converted_df['OI_R'] - converted_df['OI_R'].shift(1)) / abs(converted_df['OI_R'].shift(1)) * 100 if len(converted_df) > 0 else 0
            converted_df['EPS_QoQ'] = (converted_df['EPS'] - converted_df['EPS'].shift(1)) / abs(converted_df['EPS'].shift(1)) * 100 if len(converted_df) > 0 else 0
            converted_df['GP_R_QoQ'] = (converted_df['GP_R'] - converted_df['GP_R'].shift(1)) / abs(converted_df['GP_R'].shift(1)) * 100 if len(converted_df) > 0 else 0

            # 計算年增率兩季平均
            converted_df['Q_Rev_YoY_AVG_2Q'] = converted_df['Q_Rev_YoY'].rolling(2).mean()
            converted_df['IAT_R_YoY_AVG_2Q'] = converted_df['IAT_R_YoY'].rolling(2).mean()
            converted_df['OI_R_YoY_AVG_2Q'] = converted_df['OI_R_YoY'].rolling(2).mean()
            converted_df['EPS_YoY_AVG_2Q'] = converted_df['EPS_YoY'].rolling(2).mean()
            converted_df['GP_R_YoY_AVG_2Q'] = converted_df['GP_R_YoY'].rolling(2).mean()

            # 計算年增率兩季平均的增減
            converted_df['Q_Rev_YoY_AVG_2Q_Diff'] = converted_df['Q_Rev_YoY_AVG_2Q'] - converted_df['Q_Rev_YoY_AVG_2Q'].shift(1)
            converted_df['IAT_R_YoY_AVG_2Q_Diff'] = converted_df['IAT_R_YoY_AVG_2Q'] - converted_df['IAT_R_YoY_AVG_2Q'].shift(1)
            converted_df['OI_R_YoY_AVG_2Q_Diff'] = converted_df['OI_R_YoY_AVG_2Q'] - converted_df['OI_R_YoY_AVG_2Q'].shift(1)
            converted_df['EPS_YoY_AVG_2Q_Diff'] = converted_df['EPS_YoY_AVG_2Q'] - converted_df['EPS_YoY_AVG_2Q'].shift(1)
            converted_df['GP_R_YoY_AVG_2Q_Diff'] = converted_df['GP_R_YoY_AVG_2Q'] - converted_df['GP_R_YoY_AVG_2Q'].shift(1)

    except Exception as e:
        print(file, e)
        continue
    
    template_df = pd.concat([template_df, converted_df], ignore_index=True, join='outer')
    print(file, "success")

# 儲存結果
template_df.to_excel(output_file, index=False)
```
