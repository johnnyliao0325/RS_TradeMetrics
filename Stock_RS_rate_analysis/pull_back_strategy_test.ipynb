```python
import requests
import pandas as pd
import numpy as np
import warnings
import datetime
warnings.filterwarnings('ignore')
# ============上市股票df============
url = "https://isin.twse.com.tw/isin/class_main.jsp?owncode=&stockname=&isincode=&market=1&issuetype=1&industry_code=&Page=1&chklike=Y"
response = requests.get(url)
listed = pd.read_html(response.text)[0]
listed.columns = listed.iloc[0,:]
listed = listed[["有價證券代號","有價證券名稱","市場別","產業別","公開發行/上市(櫃)/發行日"]]
listed = listed.iloc[1:]

# ============上櫃股票df============
urlTWO = "https://isin.twse.com.tw/isin/class_main.jsp?owncode=&stockname=&isincode=&market=2&issuetype=&industry_code=&Page=1&chklike=Y"
response = requests.get(urlTWO)
listedTWO = pd.read_html(response.text)[0]
listedTWO.columns = listedTWO.iloc[0,:]
listedTWO = listedTWO.loc[listedTWO['有價證券別'] == '股票']
listedTWO = listedTWO[["有價證券代號","有價證券名稱","市場別","產業別","公開發行/上市(櫃)/發行日"]]

# ============上市股票代號+.TW============
stock_1 = listed["有價證券代號"]
stock_num = stock_1.apply(lambda x: str(x) + ".TW")
stock_num.loc[len(stock_num)+1] = '0050.TW'
stock_num.loc[len(stock_num)+1] = '^TWII'
# print(stock_num)

# ============上櫃股票代號+.TWO============
stock_2 = listedTWO["有價證券代號"]
stock_num2 = stock_2.apply(lambda x: str(x) + ".TWO")
# print(stock_num2)

# ============concate全部股票代號============
stock_num = pd.concat([stock_num, stock_num2], ignore_index=True)
# print(stock_num)
allstock_info = pd.concat([listed, listedTWO], ignore_index=True)
allstock_info.columns = ["ID","有價證券名稱","市場別","產業別","公開發行/上市(櫃)/發行日"]
allstock_info.set_index('ID', inplace = True)
print(allstock_info)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from datetime import datetime

# Load the data
symbol_csv_list = []
base_path = r'C:\Users\User\Desktop\stock\data_test'
for symbol in stock_num:
    try:
        csv_path = os.path.join(base_path, f'{symbol}.csv')
        symbol_csv = pd.read_csv(csv_path, parse_dates=['Date'])
        symbol_csv['Date'] = symbol_csv['Date'].map(lambda x: datetime.strftime(x, '%Y-%m-%d'))
        # symbol_csv.set_index('Date', inplace=True)
        symbol_csv = symbol_csv.iloc[:]
        symbol_csv.reset_index(drop=True, inplace=True)
        symbol_csv_list.append(symbol_csv)
    except:
        pass
```

## 拉回到MA 10天後漲跌

```python

rocp_filter_df = pd.DataFrame()
# rocp_filter_df['ID'] = [symbol_csv['ID'][0] for symbol_csv in symbol_csv_list]
# rocp_filter_df.set_index('ID', inplace=True)
ma_filter_dict = {}
for symbol_csv in symbol_csv_list[:]:
    symbol_csv['Adj High'] = symbol_csv['High'] * symbol_csv['Adj Close'] / symbol_csv['Close']
    symbol_csv['Adj Low'] = symbol_csv['Low'] * symbol_csv['Adj Close'] / symbol_csv['Close']
    symbol_csv['Adj Open'] = symbol_csv['Open'] * symbol_csv['Adj Close'] / symbol_csv['Close']
    symbol_csv_dorp15 = symbol_csv.iloc[:-30]
    # symbol_csv.reset_index(drop=True, inplace=True)
    try:
        # N MA 介於當天最高、最低價格之間
        for j in ['5','10','20','50']:
            ## filter 1: 5MA>20MA>50MA>200MA ## filter 2: Adj High > N MA > Adj Low
            try:
                ma_filter = lambda x: (x['20MA'] > x['50MA']) and (x['50MA'] > x['200MA']) and (x['5MA'] > x['20MA']) and (x['Adj High'] > x[j+'MA']) and (x[j+'MA'] > x['Adj Low'])
                
                pull_back_index = symbol_csv_dorp15[symbol_csv_dorp15.apply(ma_filter, axis=1)].index.values
                pull_back_index_after_10Days = pull_back_index + 10
                pull_back_index_after_20Days = pull_back_index + 20
                # print(symbol_csv.loc[pull_back_index, 'Date'])
                # print(symbol_csv.loc[pull_back_index_after_10Days, 'Date'])
                # print(symbol_csv.loc[pull_back_index, 'Date'], symbol_csv.loc[pull_back_index, 'ID'])
                price_diff_after_10Days = list(map(lambda x,y: 100*(x-y)/y, symbol_csv.loc[pull_back_index_after_10Days, 'Adj Close'], symbol_csv.loc[pull_back_index, 'Adj Close']))
                price_diff_after_20Days = list(map(lambda x,y: 100*(x-y)/y, symbol_csv.loc[pull_back_index_after_20Days, 'Adj Close'], symbol_csv.loc[pull_back_index, 'Adj Close']))
            except Exception as e:
                print(e)
                continue
            try:
                ma_filter_dict[f'mean ROCP after 10 Days'].append(np.mean(price_diff_after_10Days))
                ma_filter_dict[f'mean ROCP after 20 Days'].append(np.mean(price_diff_after_20Days))
            except Exception as e:
                ma_filter_dict[f'mean ROCP after 10 Days'] = []
                ma_filter_dict[f'mean ROCP after 10 Days'].append(np.mean(price_diff_after_10Days))
                ma_filter_dict[f'mean ROCP after 20 Days'] = []
                ma_filter_dict[f'mean ROCP after 20 Days'].append(np.mean(price_diff_after_20Days))
            try:
                ma_filter_dict[f'pullback to N_MA'].append(j)
            except Exception as e:
                ma_filter_dict[f'pullback to N_MA'] = []
                ma_filter_dict[f'pullback to N_MA'].append(j)
            
            try:
                for i in range(9):
                    rs_filter = lambda x: (x > 10+10*i) and (x < 10+10*(i+1))
                    pull_back_index_with_highRS = symbol_csv_dorp15[symbol_csv_dorp15.apply(ma_filter, axis=1)].loc[symbol_csv_dorp15['ERS_rate_250'].map(rs_filter)].index.values
                    pull_back_index_after_10Days_with_highRS = pull_back_index_with_highRS + 10
                    pull_back_index_after_20Days_with_highRS = pull_back_index_with_highRS + 20
                    # print(symbol_csv.loc[pull_back_index_with_highRS, 'Date'])
                    # print(symbol_csv.loc[pull_back_index_after_10Days_with_highRS, 'Date'])
                    # print(symbol_csv.loc[pull_back_index_after_20Days_with_highRS, 'Date'])
                    price_diff_after_10Days_with_highRS = list(map(lambda x,y: 100*(x-y)/y, symbol_csv.loc[pull_back_index_after_10Days_with_highRS, 'Adj Close'], symbol_csv.loc[pull_back_index_with_highRS, 'Adj Close']))
                    price_diff_after_20Days_with_highRS = list(map(lambda x,y: 100*(x-y)/y, symbol_csv.loc[pull_back_index_after_20Days_with_highRS, 'Adj Close'], symbol_csv.loc[pull_back_index_with_highRS, 'Adj Close']))
                    # if len(price_diff_after_10Days_with_highRS) > 0:
                    #     try:
                    #         ma_filter_dict[f'mean ROCP after 10 Days with high RS_{10+10*i}~{10+10*(i+1)}'].append(np.mean(price_diff_after_10Days_with_highRS))
                    #     except Exception as e:
                    #         print(e)
                    #         ma_filter_dict[f'mean ROCP after 10 Days with high RS_{10+10*i}~{10+10*(i+1)}'] = []
                    #         ma_filter_dict[f'mean ROCP after 10 Days with high RS_{10+10*i}~{10+10*(i+1)}'].append(np.mean(price_diff_after_10Days_with_highRS))
                    # else:
                    #     try:
                    #         ma_filter_dict[f'mean ROCP after 10 Days with high RS_{10+10*i}~{10+10*(i+1)}'].append(0)
                        # except Exception as e:
                        #     print(e)
                        #     ma_filter_dict[f'mean ROCP after 10 Days with high RS_{10+10*i}~{10+10*(i+1)}'] = []
                        #     ma_filter_dict[f'mean ROCP after 10 Days with high RS_{10+10*i}~{10+10*(i+1)}'].append(0)
                    if len(price_diff_after_20Days_with_highRS) > 0:
                        try:
                            ma_filter_dict[f'mean ROCP after 20 Days with high RS_{10+10*i}~{10+10*(i+1)}'].append(np.mean(price_diff_after_20Days_with_highRS))
                        except Exception as e:
                            print(e)
                            ma_filter_dict[f'mean ROCP after 20 Days with high RS_{10+10*i}~{10+10*(i+1)}'] = []
                            ma_filter_dict[f'mean ROCP after 20 Days with high RS_{10+10*i}~{10+10*(i+1)}'].append(np.mean(price_diff_after_20Days_with_highRS))
                    else:
                        try:
                            ma_filter_dict[f'mean ROCP after 20 Days with high RS_{10+10*i}~{10+10*(i+1)}'].append(0)
                        except Exception as e:
                            print(e)
                            ma_filter_dict[f'mean ROCP after 20 Days with high RS_{10+10*i}~{10+10*(i+1)}'] = []
                            ma_filter_dict[f'mean ROCP after 20 Days with high RS_{10+10*i}~{10+10*(i+1)}'].append(0)
            except Exception as e:
                ma_filter_dict['pullback to N_MA'].pop()
                ma_filter_dict['mean ROCP after 10 Days'].pop()


    except Exception as e:
        print(e)
        pass
ma_filter_dict_df = pd.DataFrame(ma_filter_dict)
group_by_pullback_to_N_MA = ma_filter_dict_df.groupby('pullback to N_MA')
# group_by_rs_range = rocp_filter_df.groupby('RS range')
# print(group_by_range.mean())
# print(group_by_range.count())
MA_described = group_by_pullback_to_N_MA.describe()
# RS_described = group_by_rs_range.describe()
```
